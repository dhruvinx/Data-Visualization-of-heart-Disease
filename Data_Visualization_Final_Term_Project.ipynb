```python
#Replacing the value
heart['sex'] =  heart['sex'].replace({1: 'Male', 0: 'Female'})
heart['cp'] =  heart['cp'].replace({0: 'typical angina', 1: 'atypical angina', 2: 'non-anginal pain', 3: 'asymptomatic'})
heart['exang'] =  heart['exang'].replace({1: 'Yes', 0: 'No'})
heart['target'] =  heart['target'].replace({1: 'Heart Disease', 0: 'No Heart Disease'})
heart
```

```python
selected_area = alt.selection_interval()
chart = alt.Chart(HeartDisease).mark_circle(size=160).encode(
            x='age:Q',
            y='trestbps:Q',
            color=alt.condition(selected_area, 'cp:N',
                     alt.value('lightgray'))
            ).properties(width=600,height=500).add_selection(selected_area)
chart
```

```python
input_dropdown = alt.binding_select(options=list(set(heart.cp)))

selected_points = alt.selection_single(fields=['cp'], bind=input_dropdown, name='Select this')

color = alt.condition(selected_points,
 alt.Color('sex:N'),
 alt.value('lightgray'))

alt.Chart(HeartDisease).mark_circle(size=50).encode(
 x='age:Q',
 y='trestbps:Q',
 color=color,
 tooltip=['age', 'sex', 'chol', 'thalach', 'target']
).add_selection(selected_points).properties(width=600,height=300)
```
